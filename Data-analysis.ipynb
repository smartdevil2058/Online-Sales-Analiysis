```python
import pandas as pd
df = pd.read_csv('Online_Sales_Data.csv')
```

```python
# data understandeing

print(df.head(3))
print(df.tail(3))
print(df.shape)
print(df.info())
print(df.describe())
print(df.columns)
```

```python
#data cleaning 

print(df.isnull().sum())
```

```python
#each region' transection and their payment method
df.groupby(['Region', 'Payment Method']).size()
```

```python
#highest unit price
max_unit_price = df.loc[df['Unit Price'].idxmax()]
print(max_unit_price)
```

```python
#highest unit sold
max_unit_sold = df.loc[df['Units Sold'].idxmax()]
print(max_unit_sold)
```

```python
# highest unit sold
max_Total_Revenue = df.loc[df['Total Revenue'].idxmax()]
print(max_Total_Revenue)
```

```python
#Average unit price
avg_price = df['Unit Price'].mean()
print(avg_price.mean())
```

```python
##Most sold Category 
filterd_category = df['Product Category'].value_counts()
print(filterd_category)
```

```python
#Top 10  Product in total Revenue
df.nlargest(10, 'Total Revenue') 
```

```python
#Region wise top revenue
df.groupby('Region')['Total Revenue'].sum().sort_values(ascending=False)
```

```python
#Cate gory and region wise 
df.groupby(['Product Category', 'Region'])['Total Revenue'].sum().sort_values(ascending=False)
```

```python
#profit according to the product name
df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False)
```

```python
#profit according to the Product Category
df.groupby('Product Category')['Profit'].sum().sort_values(ascending=False)
```

```python
#Month analysis
df['Date'] = pd.to_datetime(df['Date'])
df.groupby(df['Date'].dt.month)['Total Revenue'].sum().sort_values(ascending=False)
```

```python
#total revenue and profit
Total_revenue = df['Total Revenue'].sum()
print(Total_revenue)
Total_profit = df['Profit'].sum()
print(Total_profit)
```
